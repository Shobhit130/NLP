```python
#scrape text from a url (scrape web)
from urllib import request
```

```python
url = "https://www.gutenberg.org/files/98/98-0.txt"
```

```python
response = request.urlopen(url)
```

```python
raw = response.read().decode('utf8')
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
```

```python
print(tokens[:20])
```

```python
#BeautifulSoup useful for extraction. has lot of language support.
#Preprocessing --> RE to clean any html tags or unnecessary char sequence
#PoS tagging
```

```python
#Stemming
from nltk.stem import PorterStemmer
```

```python
porter = PorterStemmer()
porter.stem('joyous')
porter.stem('cacti')
```

```python
#Stemming
from nltk.stem import LancasterStemmer
```

```python
lancaster = LancasterStemmer()
lancaster.stem('joyous')
lancaster.stem('cacti')
```

```python
#Stemming
from nltk.stem import RegexpStemmer
```

```python
regexp = RegexpStemmer("ing")
regexp.stem("walking")
```

```python
#Stemming
from nltk.stem import SnowballStemmer
```

```python
snowball = SnowballStemmer("spanish")
snowball.stem("walking")
snowball.stem("cacti")
snowball.stem("mexicanos")
```

```python
for token in tokens:
    print(porter.stem(token))
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
print(lemma.lemmatize("am"))
print(lemma.lemmatize("am",pos = 'v'))
print(lemma.lemmatize("mice"))
print(lemma.lemmatize("good"))
print(lemma.lemmatize("better"))
print(lemma.lemmatize("best"))
print(lemma.lemmatize("cacti"))
```
